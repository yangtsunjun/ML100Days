```python
# - 1.
print('Filtered by first column > 20')
print(df[ df[0] > 20 ])

# - 2. 
print('\n', 'Filtered by first column + second column > 50')
print(df[ (df[0]+df[1]) > 50 ])

# - 3. 
print('\n', 'Filtered by first column < 30 or second column > 30')
print(df[ (df[0]<30)|(df[1]>30) ])

# - 4. 
print('\n', 'Filtered by total sum of row > 100')
print(df[ (df[0]+df[1]+df[2]+df[3]) > 100 ])
```
